```python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("SIKU-BERT/sikuroberta")

model = AutoModel.from_pretrained("SIKU-BERT/sikuroberta")

tokenizer.save_pretrained("./sikuberta_tokenizer")
```

```python
sen = "汉兴，接秦之坏，丈夫从军旅，老弱转粮饷，作业剧而财匮，自天子不能具钧驷，而将相或乘牛车，齐民无藏盖。"
tokens = tokenizer.tokenize(sen)
ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
ids = tokenizer.encode(sen, add_special_tokens=True)
ids
```

```python
str_sen = tokenizer.decode(ids, skip_special_tokens=False)
str_sen
```
